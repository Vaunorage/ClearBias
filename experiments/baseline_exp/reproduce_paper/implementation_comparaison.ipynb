Dans le premier volet de notre travail nous évaluons les méthodes de découverte de discrimination.

Les méthodes sont évaluées sur les métriques suivantes :
-	TSN (Total Sample Number) : Le nombre total d'échantillons générés ou testés lors d'une expérience de test d'équité algorithmique.
-	DSN (Discriminatory Sample Number) : Le nombre d'échantillons discriminatoires détectés parmi les échantillons testés.
-	DSS (Discriminatory Sample Search time) : Le temps moyen nécessaire pour générer un échantillon discriminatoire, calculé généralement en divisant le temps total par le nombre d'échantillons discriminatoires trouvés.
-	SUR (Success Rate) : Le taux de réussite dans la génération d'échantillons discriminatoires, calculé en divisant le nombre d'échantillons discriminatoires (DSN) par le nombre total d'échantillons générés (TSN).

Les méthodes évaluées sont :
-	Aequitas : Automated directed fairness testing (2018) 
-	EXPGA : Explanation-Guided Fairness Testing through Genetic Algorithm (2022) 
-	SG : Black box fairness testing of machine learning models (2019)
-	ADF : White-box fairness testing through adversarial sampling (2020)

```python
from path import HERE
from pathlib import Path
import sqlite3
import pandas as pd

DB_PATH = Path(HERE.joinpath("experiments/baseline_exp/exp.db"))

conn = sqlite3.connect(DB_PATH)
sql = f"""
SELECT * 
FROM main.paper_reproduction_results
"""
experiments = pd.read_sql(sql, conn)

experiments
```

```python
# Replace values where condition is met
experiments.loc[experiments['dataset'] == 'adult', 'dataset'] = 'Census'
experiments.loc[experiments['dataset'] == 'credit', 'dataset'] = 'Credit'
experiments.loc[experiments['dataset'] == 'bank', 'dataset'] = 'Bank'

experiments.loc[experiments['algorithm'] == 'EXPGA', 'algorithm'] = 'ExpGA'
experiments.loc[experiments['algorithm'] == 'EXPGA', 'algorithm'] = 'ExpGA'

experiments.loc[(experiments['feature'] == 'Attr8_T') & (experiments['dataset'] == 'Census'), 'feature'] = 'gender'
experiments.loc[(experiments['feature'] == 'Attr9_T') & (experiments['dataset'] == 'Census'), 'feature'] = 'race'

experiments.loc[(experiments['feature'] == 'Attr8_T') & (experiments['dataset'] == 'Credit'), 'feature'] = 'gender'
experiments.loc[(experiments['feature'] == 'Attr12_T') & (experiments['dataset'] == 'Credit'), 'feature'] = 'age'

experiments.loc[(experiments['feature'] == 'Attr2_T') & (experiments['dataset'] == 'Bank'), 'feature'] = 'marital'
experiments.loc[(experiments['feature'] == 'Attr3_T') & (experiments['dataset'] == 'Bank'), 'feature'] = 'education'
```

```python
results = []
for i, el in reference_data.iterrows():
    key = ((experiments['feature'] == el['feature']) &
           (experiments['dataset'] == el['dataset']) &
           (experiments['model'] == el['model']) &
           (experiments['algorithm'] == el['algorithm']))
    res = pd.DataFrame(
        {'feature': [el['feature']], 'dataset': [el['dataset']], 'model': [el['model']],
         'algorithm': [el['algorithm']]})
    if key.any():
        experiments_df = experiments[key][['TSN', 'DSN', 'DSS', 'SUR']].mean()
        res['ref_TSN'] = el['TSN']
        res['TSN'] = experiments_df['TSN']
        res['ref_DSN'] = el['DSN']
        res['DSN'] = experiments_df['DSN']
        res['ref_DSS'] = el['DSS']
        res['DSS'] = experiments_df['DSS']
        res['ref_SUR'] = el['SUR']
        res['SUR'] = experiments_df['SUR']
    else:
        res['ref_TSN'] = None
        res['TSN'] = None
        res['ref_DSN'] = None
        res['DSN'] = None
        res['ref_DSS'] = None
        res['DSS'] = None
        res['ref_SUR'] = None
        res['SUR'] = None
    results.append(res)
    
results = pd.concat(results)
results
```
